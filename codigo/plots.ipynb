## PARA PCA

```python
from joblib import load, dump
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

cv_estimator = load('../estimators/cv_estimator__pca_50_300_25_full')
df = pd.DataFrame(cv_estimator.cv_results_)
df_plt = df[['param_regressor__pca__n_components', 'param_regressor__reg__alpha', 'mean_test_score']]
```

```python
idx = df_plt['param_regressor__reg__alpha'] == 10000

plt.plot(df_plt[idx]['param_regressor__pca__n_components'], -df_plt[idx]['mean_test_score'], '-o', label='alpha')
plt.xlabel('Nº Componentes', fontsize=20)
plt.ylabel('MAE', fontsize=20)

plt.legend()
plt.title('MAE - Nº Componentes PCA Ridge', fontsize=20)
plt.show()

print(f'Mejor estimador: {cv_estimator.best_estimator_}')
print(f'Mejor score: {max(df["mean_test_score"])}')
```

```python
idx = df_plt['param_regressor__pca__n_components'] == 275

plt.xscale('log')
plt.plot(df_plt[idx]['param_regressor__reg__alpha'], -df_plt[idx]['mean_test_score'], '-o', label='alpha')
plt.xlabel('alpha', fontsize=20)
plt.ylabel('MAE', fontsize=20)

plt.legend()
plt.title('MAE - Alpha PCA', fontsize=20)
plt.show()

print(f'Mejor estimador: {cv_estimator.best_estimator_}')
print(f'Mejor score: {max(df["mean_test_score"])}')
```

## Comparar modelos

```python
cv_estimator_pls = load('../estimators/cv_estimator__pls_1_20_1_full_tg_200_100n')
cv_estimator_mlp = load('../estimators/cv_estimator__mlp_50_200_25_full_tg_200_100n')
cv_estimator_pca = load('../estimators/cv_estimator__pca_50_300_25_full')
```

```python
from joblib import load, dump
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data_2018 = pd.read_csv('../datasets/data_stv_2018.csv', index_col=0, parse_dates=True)
target_2018 = pd.read_csv('../datasets/target_stv_2018.csv', index_col=0, parse_dates=True, header=None)
fullData_2017 = pd.read_csv('../datasets/data_target_stv_2017.csv', index_col=0, parse_dates=True)
x_col, target_col = fullData_2017.columns[:-1], fullData_2017.columns[-1]
data_2017, target_2017 = fullData_2017[x_col], fullData_2017[target_col]
# Entre 0 y 1
target_2018 = target_2018 / 17560
```

```python
from sklearn.metrics import mean_absolute_error
y_pred = np.clip(cv_estimator_mlp.predict(data_2018), 0., 1.)

plt.scatter(target_2018, y_pred, s=5, c='crimson');

p1 = max(max(y_pred), max(target_2017))
p2 = min(min(y_pred), min(target_2017))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.title('PCA y MLP Target Compare', fontsize=20)
plt.text(p2, p1, f'Error: {mean_absolute_error(target_2018.values, y_pred):.5f}')
plt.show()
```
