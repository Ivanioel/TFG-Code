## PARA PLS

```python
from joblib import load, dump
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#cv_estimator = load('../estimators/Sotavento/cv_estimator__pls_1_20_1_full_tg_200_100n')
cv_estimator = load('../estimators/Peninsula/cv_estimator__pls_1_20_1_full_tg_200_100n')
df = pd.DataFrame(cv_estimator.cv_results_)
plt.plot(df['param_regressor__pls__n_components'], -df['mean_test_score'], '-o')
print(f'Mejor estimador: {cv_estimator.best_estimator_}')
print(f'Mejor score: {max(df["mean_test_score"])/23000}')

plt.xlabel('Número de componentes', fontsize=20)
plt.ylabel('MAE', fontsize=20)
plt.title('MAE - NComponentes PLS', fontsize=20)
plt.xticks(list(df['param_regressor__pls__n_components'].values));
```

## PARA PCA

```python
from joblib import load, dump
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

cv_estimator = load('../estimators/Sotavento/cv_estimator__ridgePCA_150_500_25')
cv_estimator = load('../estimators/Peninsula/cv_estimator__ridgePCA_Peninsula_50_300_25')

df = pd.DataFrame(cv_estimator.cv_results_)
df_plt = df[['param_regressor__pca__n_components', 'param_regressor__reg__alpha', 'mean_test_score']]
```

```python
idx = df_plt['param_regressor__reg__alpha'] == 10000

plt.plot(df_plt[idx]['param_regressor__pca__n_components'], -df_plt[idx]['mean_test_score'], '-o', label='alpha')
plt.xlabel('Nº Componentes', fontsize=20)
plt.ylabel('MAE', fontsize=20)

plt.legend()
plt.title('MAE - Nº Componentes PCA Ridge', fontsize=20)
plt.show()

print(f'Mejor estimador: {cv_estimator.best_estimator_}')
print(f'Mejor score: {max(df["mean_test_score"])/23000}')
```

```python
idx = df_plt['param_regressor__pca__n_components'] == 450

plt.xscale('log')
plt.plot(df_plt[idx]['param_regressor__reg__alpha'], -df_plt[idx]['mean_test_score'], '-o', label='alpha')
plt.xlabel('alpha', fontsize=20)
plt.ylabel('MAE', fontsize=20)

plt.legend()
plt.title('MAE - Alpha PCA Ridge', fontsize=20)
plt.show()

print(f'Mejor estimador: {cv_estimator.best_estimator_}')
print(f'Mejor score: {max(df["mean_test_score"])}')
```
